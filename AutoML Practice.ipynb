```
import random
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, auc, confusion_matrix
import pandas as pd
import category_encoders as ce
```

```
df_2018 = pd.read_csv("/content/drive/My Drive/AutoML/data_jobs_info_2018.csv")
df_2019 = pd.read_csv("/content/drive/My Drive/AutoML/data_jobs_info_2019.csv")
```

```
df_2018.head()
```

```
input_x = df_2018.drop("job_title", axis=1)
output_y = df_2018["job_title"]
```

```
x_train, x_test, y_train, y_test = train_test_split(input_x, output_y, train_size = 0.8, test_size = 0.2)
```

##Save the splitted training and testing dataset to seperate csv file

```
with open("train_data_2018","+w") as file:
  pd.concat([x_train,y_train], axis= 1).to_csv(file,index= False)
```

```
with open("test_data_2018","+w") as file:
  pd.concat([x_test, y_test], axis=1).to_csv(file,index = False)
```
