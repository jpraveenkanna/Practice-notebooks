```
"""
744. Find Smallest Letter Greater Than Target
Given a list of sorted characters letters containing only lowercase letters, and given a target letter target, find the smallest element in the list that is larger than the given target.

Letters also wrap around. For example, if the target is target = 'z' and letters = ['a', 'b'], the answer is 'a'.

Examples:
Input:
letters = ["c", "f", "j"]
target = "a"
Output: "c"

Input:
letters = ["c", "f", "j"]
target = "c"
Output: "f"

Input:
letters = ["c", "f", "j"]
target = "d"
Output: "f"

Input:
letters = ["c", "f", "j"]
target = "g"
Output: "j"
"""

class Solution:
    def nextGreatestLetter(self, letters: List[str], target: str) -> str:
        target = ord(target)
        if letters is None: return None
        if target >= ord(letters[-1]):
            return letters[0]
        
        startIndex = 0 
        endIndex = len(letters)
        
        while startIndex<endIndex:
            midIndex = (startIndex+endIndex)//2
            
            if ord(letters[midIndex]) <=target:
                startIndex = midIndex+1
            else:
                endIndex = midIndex
        return letters[startIndex%len(letters)]
```

```
"""
21. Merge Two Sorted Lists
Merge two sorted linked lists and return it as a new sorted list. The new list should be made by splicing together the nodes of the first two lists.
Example 1:
Input: l1 = [1,2,4], l2 = [1,3,4]
Output: [1,1,2,3,4,4]
"""

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:
        #print(l1.val,l2.val)
        if l1 is None:
            return l2
        if l2 is None:
            return l1
        if l1.val <= l2.val:
            #print("l1",l1.val)
            l1.next = self.mergeTwoLists(l1.next,l2)
            return l1
        else:
            #print("l2",l2.val)
            l2.next = self.mergeTwoLists(l1,l2.next)
            return l2
```

```
"""
83. Remove Duplicates from Sorted List

Given a sorted linked list, delete all duplicates such that each element appear only once.

Example 1:

Input: 1->1->2
Output: 1->2
Example 2:

Input: 1->1->2->3->3
Output: 1->2->3
"""

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteDuplicates(self, head: ListNode) -> ListNode:
        if head is None: return None
        dummyHead = ListNode("a")
        dummyHead.next = head
        
        def removeNode(prev,node):
            if node:
                if prev.val == node.val:
                    prev.next = node.next
                    removeNode(prev,node.next)
                else:
                    removeNode(node,node.next)
            else:
                return None
        removeNode(dummyHead,head)
        return dummyHead.next
```

```
"""
234. Palindrome Linked List

Given a singly linked list, determine if it is a palindrome.

Example 1:

Input: 1->2 -> 1
Output: false
Example 2:

Input: 1->2->2->1
Output: true

"""

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def isPalindrome(self, head: ListNode) -> bool:
        actual = []
        current = head
        while current:
            actual.append(current.val)
            current = current.next
        for first, last in zip(actual,actual[::-1]):
            if first != last: return False
        return True



# two pointer

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

# 1 3 4 5 4 3 1
# 1 4 4 1
# 1 3 4 5
class Solution:
    def isPalindrome(self, head: ListNode) -> bool:
        if not head: return True
        
        slow = head
        fast = head
        half_list = []
        even = False

        while slow:
            half_list.append(slow.val)
                
            if fast.next and fast.next.next:
                
                fast = fast.next.next
                slow = slow.next
                
            else:
                if fast.next: even = True
                break
        
        if not even:
            half_list.pop()

        slow = slow.next
        while slow:
            if slow.val != half_list.pop():
                return False
            slow = slow.next
        return True
            
        
        
```

```
#Given a sorted array nums, remove the duplicates in-place 
#such that each element appear only once and return the new length.
#Do not allocate extra space for another array, 
#you must do this by modifying the input array in-place with O(1) extra memory.

def removeDuplicates(nums,count = 0):
    if len(nums) == 0:
        return 0
    
    prev_val = nums[0]
    for current_val in nums[1:]:
        if current_val == prev_val:
            nums.remove(current_val)
            count +=1
        prev_val = current_val
    return len(nums),count
            
removeDuplicates([1, 2, 3, 3, 4])
```

```
def twoSum(nums,target):
  prevMap = {} # val : index

  for i, n in enumerate(nums):
      diff = target - n
      if diff in prevMap:
          return [prevMap[diff], i]
      prevMap[n] = i
  return
twoSum([2,7,11,15], 9)
```

```
class node:
  def __init__(self,value):
    self.data = value
    self.left = None
    self.right = None
class bst:
  def __init__(self):
    self.root = None

  def insert(self,val):
    current_node = self.root
    new_node = node(val)
    if current_node is None:
      self.root = new_node
      print("adding root",val)
    else:
      while current_node:
      
        if current_node.data > val:
          if current_node.left is None:
            current_node.left = new_node
            print("inserting left",current_node.data, val)
            return
          else:
            current_node = current_node.left
        elif current_node.data <= val:
          if current_node.right is None:
            current_node.right = new_node
            print("inserting right",current_node.data, val)
            return
          else:
            current_node = current_node.right


if __name__ == '__main__':
  b = bst()
  b.insert(5)
  b.insert(2)
  b.insert(6)
  b.insert(5)
  b.insert(7)
  b.insert(6)
  b.insert(8)
  b.insert(1)
  """
  root: 5
inserting 2 to left of 5
inserting 6 to right of 5
inserting 5 to left of 6
inserting 7 to right of 6
inserting 6 to left of 7
inserting 8 to right of 7
inserting 1 to left of 2
""
```
